New Hypotheis Formulation 
> H0 (NULL Hypothesis):- The average output must increse by $5000 i.e. Mean difference MD =(M1-M2)<=5000

>H1 (Alternate Hypothesis):- The average output must increase by $5000 i.e. Mean difference Md(M1-M2)>5000

> Type 1 error rejecting NULL hypothesis when it is true 

> Probability of Type 1 effor is significant level or alpha, which is given as 0.05 or 5%

> NUllHypothesis will be rejected if t_statistic > t_critical

> t_critical = 1.69 for 0.05 significance level and degree of freedom is (sample size-1) i.e.29

```
t_critical = stats.t.ppf(1-0.05, 29)
print (t_critical)
```
